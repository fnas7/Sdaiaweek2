```python
from sklearn.preprocessing import LabelEncoder
label_encoder  = LabelEncoder
df_sal['Education Level'] = LabelEncoder().fit_transform(df_sal['Education Level'])
df_sal['Job Title'] = LabelEncoder().fit_transform(df_sal['Job Title'])
df_sal['Gender'] = LabelEncoder().fit_transform(df_sal['Gender'])
df_sal.head()
```

```python
x = df_sal.drop('Salary', axis=1)
y = df_sal['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
print('Mean squared error: ', mean_squared_error(y_test, y_pred))
print('Mean absolute error: ', mean_absolute_error(y_test, y_pred))
print('R2 score: ', r2_score(y_test, y_pred))
print('Mean absolute percentage error: ', mean_absolute_percentage_error(y_test, y_pred))
```

```python
lr = LinearRegression()
Lr = lr.fit(x_train, y_train)
Lr.score(x_test, y_test)
```

```python
import plotly.express as px
from plotly.offline import iplot

df_plot = pd.DataFrame({
    'Actual_Salary': y_test,
    'Predicted_Salary': y_pred,
    'error': y_test - y_pred  # Calculate the error
})

# Create the scatter plot
fig = px.scatter(
    df_plot,
    x='Actual_Salary',
    y='Predicted_Salary',
    color='error',
    opacity=0.8,
    title='Predicted Vs. Actual',
    template='plotly_dark',
    trendline='ols'
)


fig.update_layout(
    title = {
        "font" :{
            "size" : 28,
            "family" : "tahoma"
        }
    }
)
iplot(fig)
```
